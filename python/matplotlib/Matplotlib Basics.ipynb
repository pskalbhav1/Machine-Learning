```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plt.show() is used if not using jupyter notebook
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1]) 
plt.plot(x, y, 'b') # 'r' is the color red
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
fig=plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig=plt.figure()
ax1 = fig.add_axes([0,0,1,1])
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax2.set_xlim([0,100])
ax2.set_ylim([0,200])
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
fig=plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig=plt.figure()
ax1 = fig.add_axes([0,0,1,1])
plt.plot(x, z, 'b') # 'r' is the color red
plt.xlabel('x')
plt.ylabel('z')
plt.title('title')
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
plt.plot(x, y, 'b') # 'r' is the color red
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
for ax in axes:
    axes[0].plot(x, y,'b--')
    axes[0].set_xlabel("x")
    axes[0].set_ylabel("y")
    axes[0].set_ylim([0, 200])
    axes[0].set_xlim([0,100])
    axes[0].set_title("title")

    axes[1].plot(x, z,'r')
    axes[1].set_xlabel("x")
    axes[1].set_ylabel("z")
    axes[1].set_ylim([0, 10000])
    axes[1].set_xlim([0,100])
    axes[1].set_title("title")

fig
```
